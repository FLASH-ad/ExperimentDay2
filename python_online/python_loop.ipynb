```python
#password genarator
import random
letter='abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ'
num='0123456789'
special='!@#$%^&*<>?'
all=letter+num+special
length=10
password=''
for i in range(length):
  password+=random.choice(all)
print(password)
```

```python
#swapcase
word=input('enter string : ')
res=''
for i in word:
  if i.isupper():
    res+=i.lower()
  else:
    res+=i.upper()
print(res)
```

```python
# even number -> upper
# odd number -> lower
word=input('enter string : ')
res=''
for i in range(len(word)):
  if i%2==0:
    res+=word[i].upper()
  else:
    res+=word[i].lower()
print(res)
```

```python
# remove vowel
word=input('enter a string : ')
vowel='aeiuoAEIOU'
res=''
for i in word:
  if i not in vowel:
    res+=i
print(res)
```
