```python
# check the number is prime or not
def prime(n):
    for i in range(2,n):
        if n%i==0:
            return 'not prime'

    return 'prime'
prime(77)
```

```python
# check palindrom or not
def pal(a):
    '''
    palindrom
    '''
    if a==a[::-1]:
        return 'pal'
    else:
        return 'not pal'
pal('aaai')
```
