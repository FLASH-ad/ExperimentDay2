```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
    .appName("Pyspark Fundamentals") \
    .getOrCreate()
```

```python
spark
```

```python
data = [
    ("Zaid", 66, 15,"India"),
    ("shivraj", 78,15,"India"),
    ("John", 66,50,"UK"),
    ("Sara", 97, 89,"USA")
]
columns = ["name", "age","Experience","country"]
```

```python
df = spark.createDataFrame(data, columns)
```

```python
df.show()
```

```python
df.printSchema()
```

```python
df.select("name", "age")
```

```python
df.select("name", "age").show()
```

```python
df.filter(df.age > 66).show()
```

```python
df.groupBy("country").count().show()
```
